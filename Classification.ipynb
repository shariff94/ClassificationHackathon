## Classification Problem

To distinguish between a signal process which produces Higgs bosons and a background process which does not.

`Column1` - class; Target (which needs to be predict)

`Column2 - Column22` - kinematic properties measured by the particle detectors in the accelerator


`Column23 - Column29` - features are functions of the first 21 features; these are high-level features derived by physicists to help discriminate between the two classes

### Code:

#### Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer,SimpleImputer
```

```python
# Reading the datasets
train_data = pd.read_csv("CSE7302c_train-1539668060821.csv",na_values=["?"])
test_data = pd.read_csv("CSE7302c_test-1539668060821.csv",na_values=["?"])
```

#### Analysing the train data

```python
# get the datatypes of the features
train_data.dtypes
```

From the output above, it seems that the whole dataset is taken as numeric data, lets explore to see if its true!

```python
train_data.nunique()
```

As we can see some of the features seem to categorical! And hence converting them to categorical would make prediction better

```python
# Separating numerical and categorical features
num_cols = train_data.nunique()[train_data.nunique()>5].index

cat_cols = train_data.nunique()[train_data.nunique()<=5].index
print("Numerical Columns are:",','.join(list(num_cols)),"\n\n")
print("Categorical Columns are:",','.join(list(cat_cols)))
```

```python
# Lets remove the id from the features list
num_cols = num_cols.drop("id")
num_cols
```

#### Typecasting the features in train and test datasets

```python
train_data[cat_cols] = train_data[cat_cols].astype("category")
test_data[cat_cols] = test_data[cat_cols].astype("category")
#Dropping the id column in both train and test
train_data.drop(["id"],axis=1,inplace=True)
test_data.drop(["id"],axis=1,inplace=True)
```

```python
# Checking the datatype after typecasting
train_data.dtypes
```

#### Checking for any NAs

```python
train_data.isna().sum()
```

```python
test_data.isna().sum()
```

There seems to be no missing values in train! But there are missing values in test data! and hence building a imputer model from the train data for categorical features and numerical features separately

#### Imputation of the missing data

Categorical Data imputation using SimpleImputer

```python
# Initialising the imputer object
cat_imputer = SimpleImputer(strategy="most_frequent")
```

```python
# Fitting and transforming the cateogorical data of train_data
train_data[cat_cols] = pd.DataFrame(cat_imputer.fit_transform(train_data[cat_cols]),columns=cat_cols)
train_data[cat_cols].head()
```

```python
# Imputing missing Categorical values in test dataset
test_data[cat_cols] = pd.DataFrame(cat_imputer.transform(test_data[cat_cols]),columns=cat_cols)
test_data[cat_cols].head()
```

Using KNN imputer for the numeric variables

```python
# Initialising the imputer object
num_imputer = KNNImputer()
train_data[num_cols]=pd.DataFrame(num_imputer.fit_transform(train_data[num_cols]),columns=num_cols)
train_data[num_cols].head()
```

```python
# Imputing missing Numerical values in test dataset
test_data[num_cols] = pd.DataFrame(num_imputer.transform(test_data[num_cols]),columns=num_cols)
test_data[num_cols].head()
```
